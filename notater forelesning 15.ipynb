### 1. Konsumentens preferanser

Konsumenten ønsker å oppnå en glede eller tilfredsstillelse fra konsum, noe som måles i nytte. Modellen som brukes bygger på ordinal nytteteori som sier at vi kan rangere alternative godekombinasjoner, men har ikke et presist mål på nytteenheter eller hvor intenst vi liker noe. Da kan vi ikke foreta for eksempel nyttesammenlikninger mellom konsumenter.

#### 1.1 Nyttefunksjonen

Nytten er ikke målbar i ordinal nytteteori, men vi kan bruke en funskjon for å representere konsumentens preferanser. Tenk at konsumenten får godekombinasjon A med nytte u(A)=20, eller godekombinasjon B med nytte u(B)=10. Det vi kan si er at nytten er størst med A så konsumenten velger den kombinasjonen. Vi kan ikke bruke tallene til å si at nytten av A er to ganger nytten av B. En generell nyttefunksjon som gir konsumentens preferanser over goder $x_1, x_2$ kan skrives som

$$ U = u(x_1, x_2)\tag {1}$$

Konsumentens preferanser (og dermed denne nyttefunksjonen) tenkes å ha følgende fornuftige egenskapene:      

*Fullstendig rangering* Konsumenten er i stand til å rangere alle ulike godekombinasjoner. Enten er A bedre enn B, B er bedre enn A eller konsumenten er indifferent (A og B gir like mye nytte).      
*Transitivitet* Tenk at u(A)>u(B) og u(B)>u(C). Da kan vi konkludere at A er foretrukket i forhold til C (fordi A er bedre enn B og B er bedre enn C).       
*Ikke-metning* Konsumenten foretrekker mer fremfor mindre av et gode. Gitt dette kan vi skrive følgende om **grensenytten** til godene:

$$ \frac{\partial U}{\partial x_1}=u'_1>0, \frac{\partial U}{\partial x_2}=u'_2>0\tag{2}$$

Grensenytten til gode *i* måler hvor mye nytten økes dersom konsumenten får en ekstra enhet av dette godet. Vi tenker at grensenytten avtar jo mer man har av et gode (dvs $u''_1 < 0, u''_2 <0$). Tenk at du har én time med fritid. Da vil én ekstra time med fritid verdsettes høyt. Om du har 16 timer fri vil du ikke sette så mye pris på den 17. time med fritid. Avtakende grensenytte altså.       
#### 1.2 Indifferenskurver

Nyttefunksjonen viser sammenhengen mellom tre variabler (nytte og enheter av to goder). For å kunne tegne denne funksjonen i to dimensjoner må vi fastsette en verdi for én av variablene. Vi setter en fast verdi $U_0$ på nytte og skriver
$$U_0 = u(x_1, x_2)\tag{3}$$

Likning (3) viser kombinasjoner av $x_1, x_2$ som gir nøyaktig $U_0$ i nytte. Dette er med andre ord en indifferenskurve. Se Figur 7.2b. (Legg merke til at dette likner på en produksjonsisokvant)     


La $dU_0, dx_1, dx_2$ angi små endringer i våre tre variablene. Da kan vi skrive et uttrykk for hvordan nytten endrer seg når vi endrer godekombinasjonen:
$$dU_0 = u'_1 dx_1 + u'_2 dx_2\tag{4}$$

Langs en indifferenskurve er nytte uendret slik at $dU_0=0$ og (4) skrives
$$ u'_2 dx_2= -u'_1 dx_1 \Rightarrow \frac{dx_2}{dx_1}=\frac{-u'_1}{u'_2} = MSB.$$
MSB er den marginale substitusjonsbrøken, og dette er helningen til indifferenskurven, og viser hvor villig konsumenten er til å bytte mellom godene. Se Figur 7.3.
       

Figur 7.5 viser et indifferenskart med forskjellige nyttenivå. Her antar vi at konsumenten liker begge godene slik at nytten øker i nord-østlig retning.


Legg merke til at indifferenskurvene til samme individ ikke krysser (Fig 7.6) - forklaring i forelesningen.



#### 1.3 Eksempel
Cobb-Douglas kan brukes til å illustrere preferanser:
$$ U = x_1^a x_2^b\tag {5}$$

Her er $a, b$ elastisiteter som uttrykker hvor mye konsumenten liker hvert av godene. Dersom hun får 1% ekstra av gode 1 (2) økes nytten med a% (b%). La oss tegne noen indifferenskurver.    

```python
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

```python
# symboler

x_1, x_2, a, b, U_0 =sp.symbols("x_1 x_2 a b U_0", real=True, positive=True)
U=x_1**a*x_2**b
U
```

```python
# vi må løse for x_2 og et gitt nyttenivå
x_2_ind_sol=sp.solve(U-U_0,x_2)[0]
x_2_ind_sol
```

```python
# gjør om for å plotte

indiff_x_2=sp.lambdify( (U_0,a,b,x_1), x_2_ind_sol)
indiff_x_2(U_0,a,b,x_1)
```

```python
# tegne indifferenskurver for a= .4, b= .25, U_0 = {6,8,10,12,14}

x=np.linspace(0.001,100, 100)

def create_ax():
                fig, ax = plt.subplots()
                ax.set_ylabel('$x_2$', loc='top', rotation=0)
                ax.set_xlabel('$x_1$', loc='right')
                ax.set(xlim=(0,100))
                ax.set(ylim=(0,100))
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

fig, ax = create_ax()

              

# plott funksjonen

for i in range (6, 16, 2):
    ax.plot(x, indiff_x_2(i,.4,.25,x), label='$U_{%s}$' %i)


# en annen måte å angi label med f-string
# sett "i" i {}
# trenger en til {} for å få subscript i LaTex
# siste {} følger f-string kommandoen

#for i in range(6, 16, 2):
#    ax.plot(x, indiff_x_2(i, .4, .25, x), label=f'$U_{{{i}}}$')

# tittel
ax.set_title('Indifferenskurver (a=0.4, b=.25)')

#vis navnene:
ax.legend(bbox_to_anchor=(1.5,0.6));
```

### 2. Konsumentens budsjettbetingelse

Konsumenten bruker hele sin inntekt (m) på konsum, noe som gjenspeiles i budsjettbetingelsen
$$m = p_1x_1 + p_2x_2 \tag{6}$$

hvor $p_1, p_2$ er prisen på gode 1 og 2.      

For å tegne budsjettbetingelsen kan vi omformulere slik at den står med $x_2 = .......$:

$$x_2 = \frac{m}{p_2} - \frac{p_1}{p_2}x_1 \tag{7}$$

Dette er en rett linje som forteller at konsumenten kan kjøpe maksimalt $\frac{m}{p_2}$ enheter av gode 2, og for hver enhet av gode 1 som kjøpes må konsumenten gi opp $\frac{p_1}{p_2}$ enheter av gode 2. Da er $\frac{p_1}{p_2}$ den relative prisen på gode 1, og helningen til budsjettlinjen er $-\frac{p_1}{p_2}$.        

Følgende figur viser mulighetsområdet for konsum:


```python
# definerer budsjett

def budsjett(M, p1, p2, x1):
    return M/p2 - p1*x1/p2

fig, ax = create_ax()

ax.plot(x, budsjett(2000, 25, 40, x), label = 'Konsumentens budsjett')

ax.fill_between(x, 0, budsjett(2000, 25, 40, x), color = "pink",alpha = 0.3,label='Mulighetsområdet')

ax.set_title('Konsumentens budsjettlinje og mulighetsområde')
ax.legend(loc='best');
```

### 3. Konsumentens tilpasning
Dette følger logikken for bedriftens kostnadsminimering på lang sikt (forelesning 12). Konsumenten maksimerer sin nytte gitt sitt budsjett:

$$ maksimer_{x_1,x_2} \;\; U = u(x_1, x_2) \;\; gitt\;\; m=p_1x_1+p_2x_2$$

Dette løses ved hjelp av Lagrange metoden (se Appendiks 7.1 på s. 232 i boka).         

Vi setter opp Lagrange funksjonen:
$$L = u(x_1, x_2)-\lambda (p_1x_1+p_2x_2-m)$$
hvor $\lambda$ er en Lagrange multiplikator som viser verdien av å gjøre bibetingelsen mindre streng. I dette tilfellet viser det økningen i nytten dersom vi øker intekten med 1kr. Vi deriverer $L$ med hensyn på $x_1, x_2,\lambda$:

$$\begin{align}
\frac{\partial L}{\partial x_1} &= u'_1 - \lambda p_1 = 0 \tag {8}\\
\frac{\partial L}{\partial x_2} &= u'_2 - \lambda p_2 = 0 \tag {9}\\
\frac{\partial L}{\partial \lambda} &= m - p_1x_1 - p_2x_2 = 0 \tag {10}\\
\end{align}$$

Legg merke til at (10) innebærer at løsningen som vi finner må bruke hele budsjettet. Vi kan skrive om (8) og (9):

$$\begin{align}
\frac{u'_1}{p_1} &= \lambda\tag {11}\\
\frac{u'_2}{p_2} &= \lambda \tag {12}\\
\end{align}$$

Som innebærer at

$$
\frac{u'_1}{p_1}=\frac{u'_2}{p_2} \tag{13}
$$

Dette sier at konsumentens tilpasning fordeler inntekt på en slik måte at økningen i nytte fra den siste kr brukt på gode 1 er lik økningen i nytte fra den siste kr brukt på gode 2. Vi kan skrive om (13) som følger

$$
\frac{u'_1}{u'_2}=\frac{p_1}{p_2} \tag{14}
$$


Tolkningen av (14) er at den relative prisen på gode 1 (høyre side) er lik den marginale substitusjonsbrøken (MSB).       

Å finne et maksimum for konsumenten betyr at vi må finne godekombinasjonen som er på budsjettlinjen hvor en indifferenskurve har samme helning som budsjettbetingelsen. Da må to likninger oppfylles samtidig:

$$\begin{align}
\frac{u'_1}{u'_2}=\frac{p_1}{p_2} \tag{14}\\
m=p_1x_1+p_2x_2 \tag {6}\\
\end{align}$$

(14) sikrer at kombinasjonen av $x_1, x_2$ som velges maksimerer nytten, og (6) sikrer at denne kombinasjonen bruker hele budsjettet.

#### 3.1 Cobb-Douglas preferanser

Vi skal finne etterspørsel etter gode 1 og gode 2 for en konsument som har nyttefunksjon i (5) og budsjett i (6).

```python
# nye symboler
p_1, p_2, m =sp.symbols("p_1 p_2 m", real=True, positive=True)

# budsjett
bud=p_1*x_1+p_2*x_2

# Beregne MSB

msb = sp.diff(U, x_1)/sp.diff(U,x_2)
display(msb)
```

```python
# Løs likning (14)

x_2_rel=sp.solve(msb-p_1/p_2,x_2)[0]
display(x_2_rel)
```

```python
# sett inn i m og løs for x_1

budsj=bud.subs(x_2,x_2_rel)
display(budsj)
x_1_eq= sp.solve(budsj-m,x_1)[0]

# dette er likevektsløsning for x_1
#sett inn i x_2_rel for å få x_2_eq

x_2_eq= x_2_rel.subs(x_1,x_1_eq)
display(x_1_eq)
display(x_2_eq)
```

Da har vi etterspørsel etter gode 1 og 2 som $x_1=\frac{am}{p_1(a+b)},x_2=\frac{bm}{p_2(a+b)}$. Utgiften til gode 1 og 2 blir da $p_1x_1=\frac{am}{(a+b)},p_2x_2=\frac{bm}{(a+b)}$. Vi kan beregne andelen av budsjettet brukt på hvert gode:

```python
# budsjettandeler

andel_x_1=p_1*x_1_eq/m
andel_x_2=p_2*x_2_eq/m
display(andel_x_1)
display(andel_x_2)
```

En konstant andel av inntekten brukes på hvert gode! Dersom a>b vil budsjettandelen til gode 1 være størst.

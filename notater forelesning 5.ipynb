### 1. Et eksempel

En avgift er en *indirekte skatt* som pÃ¥legges kjÃ¸per eller selger av en vare eller tjeneste. Avgiften kan ta flere former, som for eksempel en avgift per enhet omsatt eller en avgift som er en andel av salgsprisen. Ett eksempel pÃ¥ fÃ¸rstnevnte er avgiften pÃ¥ alkohol (per liter avhengig av alkoholinnhold), og ett eksempel pÃ¥ sistnevnte er merverdiavgiften. I Norge er avgifter regulert gjennom [SÃ¦ravgiftsloven](https://lovdata.no/dokument/NL/lov/1933-05-19-11) fra 1933. Her ser vi de ulike avgiftene som beregnes pÃ¥ ei flaske med vin (kilde er [Vinmonopolet](https://www.vinmonopolet.no/content/om-oss/priser-og-avgifter)):


Her skal vi se pÃ¥ virkningen av Ã¥ ilegge en vare en avgift per enhet omsatt. Betrakt fremstillingen av markedet for oppdrettslaks i lÃ¦reboka s. 101-104. EtterspÃ¸rsel er gitt ved


\begin{equation}
   x^{D}= 500 -3.2p\tag{1}
\end{equation}

og tilbud:


\begin{equation}
   x^{S}= -100 + 4.3p\tag{2}
\end{equation}

I boka ser forfatterne pÃ¥ en avgift pÃ¥ 4kr per enhet som produsenten mÃ¥ betale. Ved hjelp av Python kan vi sette opp en mer generell modell (og eksempelet i boka blir da et spesialtilfelle av det mer generelle). Vi kan selvsagt sette opp modellen i en veldig generell form ved Ã¥ bytte ut tallene i tilbud/etterspÃ¸rsel med bokstaver.     

Tenk at en avgift pÃ¥ t kr per enhet mÃ¥ betales av *selgeren*. Han tar p kr per enhet fra kjÃ¸peren og av dette mÃ¥ t kr sendes videre til myndighetene; da fÃ¥r selgeren (p-t)kr per enhet solgt.     

Dersom en avgift pÃ¥ T kr per enhet mÃ¥ betales av kjÃ¸peren, mÃ¥ han betale p kr til selgeren pluss T kr til myndighetene, altsÃ¥ (p+T) kr tilsammen. Da kan vi skrive om etterspÃ¸rsel og tilbud i (1) og (2):   



\begin{equation}
   x^{D}= 500 -3.2(p+T)\tag{3}
\end{equation}

og tilbud:


\begin{equation}
   x^{S}= -100 + 4.3(p-t)\tag{4}
\end{equation}

La oss sette opp vÃ¥re funksjoner:

```python
# importer pakkene
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

```python
# EtterspÃ¸rsel i likning (3)

def x_d(p,T):
    return 500 - 3.2*(p+T)

#Tilbud i likning (4)

def x_s(p,t):
    return -100+4.3*(p-t)


```

La oss tegne eksempelet som er gitt i boka med t=4, T=0, dvs en avgift pÃ¥ 4 kr lagt pÃ¥ produsenter.

```python
p_num=np.linspace(0.1,300,100)

#lag en akse

def create_ax():
                fig, ax = plt.subplots()
                ax.set_ylabel('Pris per kilo', loc='top', rotation = 0)
                ax.set_xlabel('Tonn per uke', loc='right')
                ax.set(xlim=(220,280))
                ax.set(ylim=(70,90))
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

fig, ax = create_ax()

# plott funksjonene

ax.plot(x_d(p_num, 0),p_num, label='EtterspÃ¸rsel')

ax.plot(x_s(p_num, 0),p_num, label='Tilbud')
ax.plot(x_s(p_num, 4),p_num, label='Tilbud med avgift (t=4)')

# vis avstand mellom linjene

ax.text(250, 83.5, "4 kr")

ax.annotate("", (x_s(82.5,0),82.5), (x_s(86.5,4), 86.5), arrowprops={"arrowstyle":"<->", "color":"gray"})

# Vis likevektskvantum fÃ¸r og etter
ax.text(243.5, 79, "A")
ax.text(236, 83, "B")

# tittel
ax.set_title('Likevekt med en avgift')

#vis navnene utenfor plottet:

ax.legend(bbox_to_anchor=(1.5,0.6));
```

Den opprinnelige likevekten er i A. Den nye tilbudsfunskjonen krysser etterspÃ¸rsel i B og vi ser at omsatt mengde faller. FÃ¸r vi beregner effekten pÃ¥ likevekten, la oss spÃ¸rre hva som ville skjedd dersom vi hadde lagt avgiften pÃ¥ kjÃ¸perne (T=4)?

```python
ax.plot(x_d(p_num,4),p_num, label='EtterspÃ¸rsel med avgift (T=4)', color='blue')
ax.text(238, 78, "C")

ax.text(256, 73.5, "4 kr")

ax.annotate("", (x_d(75,0),75), (x_d(71,4), 71), arrowprops={"arrowstyle":"<->", "color":"gray"})
ax.legend(bbox_to_anchor=(1.6,0.6))
fig
```

Det ser ut som om omstatt kvantum blir det samme om vi ilegge avgiften kjÃ¸pere eller selgere! La oss lÃ¸se for likevekten og vise dette formelt.

```python
# definer symboler i sympy

x,p,t,T=sp.symbols('x p t T', positive=True, real=True)

# betingelse for likevekt

eq_cond=sp.Eq(x_d(p,T),x_s(p,t))
eq_cond
```

```python
# lÃ¸s for p
p_eq=sp.solve(eq_cond,p)
p_eq[0]
```

```python
# finn tilsvarende kvantum
x_eq=x_s(p_eq[0],t)
x_eq
```

Legg merke til at vi ser samme tall foran T (avgift lagt pÃ¥ kjÃ¸per) og t (avgift lagt pÃ¥ selger). Det er dette vi oppdaget i figuren. En avgift pÃ¥ 4 kr pÃ¥ selger eller kjÃ¸per har samme effekt pÃ¥ likevektskvantumet!

```python
# Bruk sp.lambdify for Ã¥ gjÃ¸re om sympy til numpy array

# Likevektspris som funksjon av t og T

p_eq_fn=sp.lambdify(
    (t,T), 
    p_eq[0]
)

p_eq_fn(t,T)
```

Dette ser det samme ut som fÃ¸r, men vi har nÃ¥ laget en funksjon av t og T som vi kan plotte med numpy i matplotlib.

```python
# Likevektskvantum som funksjon av t og T

x_eq_fn=sp.lambdify(
    (t,T), 
    x_eq
)

x_eq_fn(t,T)
```

```python
# Kvantum omsatt med avgift pÃ¥ selger pÃ¥ t=4
x_eq_fn(4,0)
```

```python
# Kvantum omsatt med avgift pÃ¥ kjÃ¸per pÃ¥ T=4
x_eq_fn(0,4)
```

Kvantum er det samme uansett hvem som mÃ¥ betale avgiften.

```python
x_opt = round(x_eq_fn(4,0),2)
x_opt
```

```python
# prisen som konsumenten mÃ¥ betale er
p_kons = round(p_eq_fn(4,0),2)
p_kons
```

```python
# prisen som produsenten fÃ¥r er
p_kons -4
```

```python
# sett inn linjer som viser likevekten
# clip_on=False for Ã¥ tegne utenfor plottomrÃ¥det

ax.vlines(x_eq_fn(4,0),65, 82.3, linestyles='dashed', clip_on=False)
ax.hlines(p_eq_fn(0,4),200,237, linestyles='dashed', clip_on=False)
ax.hlines(p_eq_fn(4,0),200,237, linestyles='dashed', clip_on=False)

# sett inn likevektsverdier i figuren

ax.annotate(x_opt, xy=(x_opt-4,64), annotation_clip=False)
ax.annotate(p_kons, xy=(200, p_kons), annotation_clip=False)
ax.annotate(p_kons-4, xy=(200, p_kons-4), annotation_clip=False)

ax.legend(bbox_to_anchor=(1.6,0.6))
fig
```

Forskjellen mellom disse er avgiften som betales til staten. FÃ¸r avgiften ble innfÃ¸rt var markedet i likevekt i punkt A: her betalte konsumenter `p_eq_fn(0,0)` og alt gikk til selgere.

### 3. En optimal avgiftssats?

Vi kan tenke oss at myndighetene har forskjellige mÃ¥l med avgiftspolitikken. Noen avgifter legges pÃ¥ varer som myndighetene vil at vi skal bruke mindre av, for eksempel alkohol, usunn mat og bilkjÃ¸ring. Disse gir samtidig inntekter til myndighetene. Flere ganger har [Tenk TromsÃ¸ vurdert om bompengeavgiften i TromsÃ¸ bÃ¸r Ã¸kes](https://www.itromso.no/nyheter/i/q19v3o/neppe-oekning-i-bompenger-i-tromsoe-ennaa-vi-vil-droefte-saken-denne-uken) fordi inntektene er lavere enn forventet.       

Vi kan beregne for eksempel en avgiftssats i et marked som maksimerer avgiftprovenyet. I den siste figuren ser vi at avgiftproveny er avgiften per enhet multiplisert med antall enheter solgt. La oss undersÃ¸ke dette for en avgift lagt pÃ¥ produsenten (dvs T=0). La oss fÃ¸rst undersÃ¸ke effekten som en avgift har pÃ¥ omsetningen.

```python
t_num=np.linspace(0,150,100)


fig,ax=plt.subplots()

ax.set_ylabel('Likevektsomsetning', loc = 'top', rotation = 0)
ax.set_xlabel('Avgift (kr/enhet)', loc= 'right')

ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

ax.plot(t_num, x_eq_fn(t_num, 0),label='Omsetning')



#legend:
ax.legend(loc='best', frameon=False)

#title
ax.set_title ('Effekten av en avgift pÃ¥ omsetning')

ax.set_xlim(0,150)
ax.set_ylim(0,250);
```

Myndighetene stÃ¥r overfor en *avveining* her: en Ã¸kning i avgiften gir flere kroner i proveny per enhet solgt, men reduserer samtidig antall enheter solgt. La oss undersÃ¸ke dette nÃ¦remere.

```python
def avgiftprov(t):
    return t*x_eq_fn(t,0)
avgiftprov(t)
```

```python
fig,ax=plt.subplots()

ax.set_ylabel('Proveny (kr)', loc = 'top', rotation = 0)
ax.set_xlabel('Avgift (kr/enhet)', loc= 'right')

ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

ax.plot(t_num, avgiftprov(t_num),label='Avgiftproveny')



#legend:
ax.legend(loc='best', frameon=False)

#title
ax.set_title ('Avgiftproveny')

ax.set_xlim(0,150)
ax.set_ylim(0,8500);
```

Denne type kurve kalles ofte en **Laffer kurve**. (Arthur Laffer var rÃ¥dgiver for US President Ronald Reagan, 1981-89). Her ser vi myndighetens avveining ganske tydelig, og vi kan ogsÃ¥ finne avgiftssatsen som maksimerer provenyet.

```python
# finn maks

# deriver mhp t
d_avgpr= sp.diff(avgiftprov(t),t)

# sett = 0 og lÃ¸s for t

t_opt_dict=sp.solve((d_avgpr,0),t)

t_opt = round(t_opt_dict[t],2)

t_opt

```

```python
# maksimal avgiftproveny

maks_prov = round(avgiftprov(t_opt),2)
maks_prov
```

```python
ax.hlines(maks_prov, 0, t_opt, linestyles='dashed')
ax.vlines(t_opt, 0, maks_prov, linestyles='dashed')
fig
```

En sÃ¦ravgift er en avgift som betales ved produksjon og omsetning av noen typer varer. Det kan ogsÃ¥ gjelde tollbelagte varer som innfÃ¸res fra utlandet. [Her](https://www.regjeringen.no/no/tema/okonomi-og-budsjett/skatter-og-avgifter/avgiftssatser-2025/id3057881/) finnes ei liste med avgiftssatser for 2025, hvor vi ser
at de fleste avgifter er ilagt per enhet (stykk, kilo, liter, kWh, sete osv). Noen sÃ¦ravifter er beregnet som en prosentsats, som for eksempel dokumentavgiften som mÃ¥ betales av en boligkjÃ¸per, og som utgjÃ¸r 2,5% av salgsverdien. Merverdiavgift ilegges som en prosentsats av omsetningsverdien. Den vanlige satsen i Norge er 25%, mens nÃ¦ringsmidler (mat og drikke) fÃ¥r mva 15%, og bl.a. persontransport, kinobilletter og hotellovernattinger har en sats pÃ¥ 12%.          

Her skal vi se pÃ¥ en merverdiavgift som illegges hotellovernattinger. IfÃ¸lge tabell 12892 og tabell 12897 (SSB) var det ca 187 000 hotellovernattinger i Troms og Finnmark i juli 2023 til en gjennomsnittspris pÃ¥ ca 1000kr. Tenk at etterspÃ¸rsel og tilbud er lineÃ¦re og gitt ved


\begin{equation}
   x^{D}= a - b p_k\tag{1}
\end{equation}

og tilbud:


\begin{equation}
   x^{S}= -A + B p_s\tag{2}
\end{equation}

hvor a, b, A, B er alle positive tall. KjÃ¸pere (hotellgjester) betaler $p_k$ pr rom og hotellene (selgere) fÃ¥r $p_s$. Prisen som selgere fÃ¥r er det som konsumenter betaler minus merverdiavgift mva $tp_k$, noe som betyr at $p_s=(1-t)p_k$. Her er $1>t\geq 0$ avgiftssatsen. Vi kan skrive (2) ogsÃ¥ som en funksjon pÃ¥ prisen som konsumenten betaler:

\begin{equation}
   x^{S}= -A + B p_k(1-t)\tag{3}
\end{equation}

Det fÃ¸rste vi vil gjÃ¸re er Ã¥ prÃ¸ve Ã¥ "kalibrere" modellen, dvs finne verdier pÃ¥ a, b, A, B som gir oss en likevektslÃ¸sning med p = 1000 og x = 187000, gitt at t=0.12 (mva sats for hotellovernattinger).

```python
# Definer variabler og parametre
x, p, a, b, A, B, t = sp.symbols('x p a b A B t')

# faktiske verdier
x_value = 187000  
p_value = 1000    
t_value = 0.12    

# likninger for etterspÃ¸rsel og tilbud
eq_d = sp.Eq(x, a - b * p)
eq_s = sp.Eq(x, -A + B * p * (1 - t))

# sett inn verdiene for a,p,t
eq_d_sub = eq_d.subs({x: x_value, p: p_value})
eq_s_sub = eq_s.subs({x: x_value, p: p_value, t: t_value})

# lÃ¸s likningssystem
sol = sp.solve((eq_d_sub, eq_s_sub), (a, b, A, B))

sol
```

Da fÃ¥r vi to likninger: $A=-880B-187000, a=1000b+187000$. Vi har ikke mer informasjon enn dette, sÃ¥ mÃ¥ velge parametre som oppfyller disse kravene (hadde vi hatt en priselastisitet kunne vi ogsÃ¥ ha brukt dette).        

Jeg velger a=287000, b=100, A=33000, B=250. Vi deler disse parametre pÃ¥ 1000 slik at vi mÃ¥ler antall overnattinger i tusen.

```python
# EtterspÃ¸rsel i likning (1)

def x_d(p):
    return 287 - 0.1*p

#Tilbud i likning (3)

def x_s(p,t):
    return -33+0.25*p*(1-t)
```

```python
# vi lager en figur

p_num=np.linspace(0.1,1500,100)

def create_ax():
                fig, ax = plt.subplots()
                ax.set_ylabel('Pris per natt', loc='top', rotation = 0)
                ax.set_xlabel('Antall overnattinger (tusen)', loc='right')
                ax.set(xlim=(150,220))
                ax.set(ylim=(500,1200))
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

fig, ax = create_ax()

# plott funksjonene

ax.plot(x_d(p_num),p_num, label='EtterspÃ¸rsel')

ax.plot(x_s(p_num, 0.12),p_num, label='Tilbud')

ax.hlines(1000, 150, 187, 'r', linestyles='dashed')
ax.vlines(187, 500, 1000, 'r', linestyles = 'dashed')

ax.set_title('Hotellovernattinger i Troms og Finnmark, juli 2023')

ax.legend(bbox_to_anchor=(1.5,0.6));
```
